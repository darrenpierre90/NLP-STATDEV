```python
import pandas as pd 
import numpy as np 
import os 
import pymannkendall as mk
```

```python
filePath="./data.csv"
df =pd.read_csv(filePath)
print(type(df))
numRows=len(df.index)
```

```python
def dataMatrix(df):
    yList=list()
    x= len(df)
    for _,rows in df.iterrows():
        rows=[int(rows[x].replace(",","")) for x in range(len(rows))  if x > 0]
        yList.append(rows)
        
    return yList
        
def trendDetection(rows):
    x=[index for index in range(len(rows))]
    zippedList=list()
    for x in range(len(rows)):
        result=mk.original_test(rows[x])
        zippedList.append((x,result.slope))
        print(f"this is your result \n {result} \n and this is your index {x}")
    return zippedList

test=trendDetection(dataMatrix(df))
print(test)

print(df)
```
